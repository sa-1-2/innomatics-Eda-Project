### Q1. Does Salary depends upon designation

```python
plt.figure(figsize=(15,15))
sns.boxplot(x = df3['Designation'], y = df3['log_Salary'])
plt.xticks(rotation=90)
plt.show()
```

- As per boxplot we can conclude that System analyst earn quite more than other designated candidates

Let's check whether it is statistically significant or due to chance
- Null Hypothesis: Avg Salary of each designation is equal
- Alternate Hypothesis: There is difference in salary for every designation

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
model = ols('log_Salary ~ C(Designation)', data=df3).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

As p-value is very small <0.05, we can reject null hypothesis stating that there is difference in Salaries due to designation

### Q2. Does Salary depends on Jobcity 


```python
job_cities = df3['JobCity'].value_counts().nlargest(15).index
df_top_10 = df3[df3['JobCity'].isin(job_cities)]

plt.figure(figsize=(20,15))
sns.boxplot(x=df_top_10['JobCity'], y=df_top_10['log_Salary'])
plt.xticks(rotation=90)
plt.xlabel('Salary')
plt.ylabel('Job City')
plt.title('Boxplot of Salary by Top 10 Job Cities')
plt.show()
```

- We cannot conclude by visualising boxplot that whether Jobcity play specific role in Salaries

Lets perform Hypothesis tesing using Anova
- Null Hypothesis: Avg Salary for each JobCity is equal
- Alternate Hypothesis: There is difference in avg salary for each Job city

```python
model = ols('log_Salary ~ C(JobCity)', data=df3).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

As p-value is smaller than 0.5, we can reject null hypothesis, and states that Jobcity plays a specific role in Higher/Lower Salaries

### Q3. Does male earns more than females

```python
sns.boxplot(x = df3['Gender'], y=df3['log_Salary'])
plt.show()
```

- By visualising box plot we can see there is no much difference b/w male and female salaries.

Lets perform Hypothesis testing to check 
- Null Hypothesis: Avg Salary of Male and female are equal
- Alternate Hypothesis: There is difference b/w avg salaries of Male and Female

```python
import scipy.stats as stats
m_salary = df3[df3['Gender'] == 'm']['log_Salary']
f_salary = df3[df3['Gender'] == 'f']['log_Salary']

# Perform independent samples t-test
stats.ttest_ind(m_salary, f_salary)
```

As p-value is greater than 0.05, we accept null hypothesis, stating there is not statistically significant difference of Male and female salaries
